##Parte 1: Abrindo os arquivos fits

A atividade fornecia 2 arquivos no formato FITS (Flexible Image Transport System), contendo dados ainda "desconhecidos". O primeiro passo da atividade era abrir esses arquivos e identificar seu conteúdo.

Para isso utilizaremos alguns pacotes do Python muito comuns para a manipulação de dados em astronomia, principalmente o astropy, que reúne uma série de pacotes com as rotinas mais comuns. O processo de abertura dos arquivos FITS, e sua identificação é descrito abaixo:

#Parte 2: Plotar o catálogo e os espectros

A segunda parte do exercício consistia em gerar uma imagem .png do campo do primeiro arquivo, sobreposto de um plot dos objetos do catálogo do segundo arquivo. Para isso precisamos converter as coordenadas de pixel x e y para RA e DEC, utilizando as rotinas do `wcs` do astropy.

```python
from astropy.wcs import WCS
from wcsaxes import WCSAxes

fig2 = plt.figure(figsize = (12,10))
ax2 = WCSAxes(fig2, [0.1, 0.1, 0.8, 0.8], wcs=WCS(fits1[1].header))
fig2.add_axes(ax2)

ax2.imshow(np.log(fits1[1].data), origin='lower').set_cmap('Greys')
ax2.plot(fits2[1].data['x_ccd'],fits2[1].data['y_ccd'],'wo')
for i in fits2[1].data:
    ax2.text(i['x_ccd']+15, i['y_ccd'], i['ID'])
    
fig2.savefig('campo.png')
```

Também fazia parte da tarefa plotar os espectros sobrepostos. Como os valores do comprimento de onda está em pixel e não em angstrom, devemos converter utilizando uma função especial e o wcs do astropy.

```python
fig3, ax3 = plt.subplots(1, 1, figsize=(12,10))

def pix2ang(hdu, ax):
    wcs = WCS(hdu.header)
    index = np.arange(hdu.header['NAXIS1'])
    angs = wcs.wcs_pix2world(index[:,np.newaxis], 0)
    angs = angs.flatten()
    ax.plot(angs, hdu.data, '-')
    
for i in np.array(xrange(13))+2:
    pix2ang(fits2[i],ax3)
    
fig3.savefig('spectros.png')
```

Além disso, precisamos criar um VOTable com os dados do catálogo. Isso pode ser feito facilmente com:

```python
from astropy.io.votable.tree import VOTableFile, Resource, Table, Field

# Create a new VOTable file...
votable = VOTableFile()

# ...with one resource...
resource = Resource()
votable.resources.append(resource)

# ... with one table
table = Table(votable)
resource.tables.append(table)

fields = []
for i in fits2[1].data.columns:
    fields.append(Field(votable, name=i.name, datatype=np.dtype(i.disp).name, arraysize="*"))

table.fields.extend(fields)

table.create_arrays(len(fits2[1].data))

for i in xrange(len(fits2[1].data)):
    table.array[i] = fits2[1].data[i]

votable.to_xml('catalogo.xml')
```
